## Ciclo de vida de una Botnet

La principal caracteristica de una botnet es el ciclo de vida que posee. Generalmente se divide en 4 etapas :

1. Formation
2. C&C
3. Attack
4. Post-Attack





### 1. Formation

La formacion es el paso inicial del sistema. La botnet se propaga a traves de dos mecanismos :

1. Infeccion por vulnerabilidad

2. Infeccion por descarga 


En la **infeccion por vulnerabilidad**, la botnet tiene la capacidad para realizar un **inbound scan** o **exploracion entrante**. Esto le permitr explorar y descubrir vulnerabilidades del target, los cuales se intentan explotar de manera que se pueda comprometer al objetivo. Una vez dentro del target, el sistema es capaz de descargar un codigo malicioso proveniente de un server administrado por el Bot-Master
 
La **infeccion por descarga**, es la forma mas comun de infeccion. Es provocada por el desconocimiento del usuario, ya que se hace evidente el uso de ingenieria social para hacer uso indiscriminado de URL maliciosas, las cuales permiten descargar el codigo binario a partir de **bugs del navegador, malas configuraciones** y **controles inseguros en lenguajes de programacion (ejemplo : Javascript)**

### 2. C&C ( Command and Control)

Es la segunda etapa de la botnet, en la cual el Bot-Master es capaz de manipular mediante una serie de comandos a los "zombies" o dispositivos infectados. Esta etapa depende mucho de la jerarquia que se tenga predeterminada para el sistema, siendo esta :


1. Centralizada 
2. Descentralizada (P2P)

<img src="./Resources Proyecto de grado/jerarquia.png">


#### 2.1 Centralizada

Son aquellos sistemas que manejan protocolos tales como **IRC (Internet Relay Chat)** y **HTTP** para sus comunicaciones. Los bots o zombies son controlados mediante un solo servidor central. La desventaja es que si el servidor es rastreado y encontrado fisicamente (mediante su direccion IP), es muy posible que la botnet entera sea dada de baja.

#### 2.2 Descentralizada

Los cyber-delincuentes se vieron en aprietos una vez la jerarquia centralizada mostro sus desventajas. Estos, decidieron actuar y encontraron la forma de hacer mas dificil el trabajo de los investigadores de seguridad informatica. Ellos, decidieron descentralizar las comunicaciones que se llevaban a cabo en todo el sistema, e implementaron un modelo semejante al **P2P**, donde fuese mucho mas dificil encontrar la ubicacion fisica del kernel de la botnet. La comunicacion se hace entre nodos, de tal manera que si se quiere llegar al servidor, se tendra que hacer un barrido por toda la red de nodos presentes en la botnet ( teniendo en cuenta que pueden ser millones de dispositivos involucrados!)

### 3. Attack 

El cyber-delincuente se encarga de realizar el ataque. Para ello, debera actualizar la direccion (IP por ejemplo ) a la cual desea dirigir el ataque. Una vez realizado esto, mediante la consola de C&C, ordena a los zombies actualizar los nuevos parametros introducidos, donde una vez configurada toda la red, comienza el ataque....


Se comienza con un ataque DDoS, donde se envian gran cantidad de solicitudes al target. El servidor no logra balancear la cantidad de paquetes entrantes, por lo que empieza a saturarse, de manera que queda inhabilitado por un tiempo limitado, mientras que los ingenierios de seguridad restablecen los servidores y la red entera.

<img src="./Resources Proyecto de grado/ddos.png">

### 4. POST-Attack

El post-Attack es la etapa donde el cyber-delincuente reconfigura toda la red, esperando no haber dejado rastros en el ataque. Tambien se asegura de revisar si los zombies siguen activos o si sus ips fueron descubiertas en algun momento.

## Mecanismos de mitigacion de Botnet

### 1. Pasivas

#### 1.1 Honeypots

Los **Honeypots** son recursos informaticos estrechamente monitoreados cuyo objetivo es tentar al atacante a realizar un ataque, con el fin de realizar investigaciones sobre sistemas de seguridad en la actualidad.

El unico problemas es que en algunos paises tienen un uso no autorizado o es ilegal e uso de esta clase de herramientas.

Estos sistemas pueden ejecutar ciertos componentes y de esto depende el tipo de interaccion entre ellos, los cuales pueden ser de dos tipos:


1. **Alta Interaccion** : Utilizan recursos reales como sistemas operativos y aplicaciones.

2. **Baja Interaccion** : Permiten emular servicios

<img src="./Resources Proyecto de grado/hoeypot.png">

#### 1.2. Honeynets

Son redes de seguridad utilizadas para capturar actividad maliciosa y pueden poseer 2 o mas **honeypots**. Cualquier interaccion con una honeynet puede ser considerada como maliciosa

<img src="./Resources Proyecto de grado/honeynet.png">

#### 1.3 Honeywall

Son mecanismos que poseen dos capas :

1. **Capa de control de datos**: Permite mitigar los ataques a los sistemas que se estan exponiendo como "señuelo".

2. **Capa de captura de datos** : La cual permite monitorear las actividades que se estan llevando a cabo en la red.

<img src="./Resources Proyecto de grado/honeywall.png">

#### 1.4 BotHunter ( Inspeccion de paquetes)

Son herramientas diseñadas para la deteccion de botnets. Estan basadas en un algoritmo llamado **Correlacion de dialogo de red**. 

Fueron desarollados bajo el programa **Cyber-TA** en el laboratorio de **Ciencias Computacionales en SRI Internacional**

Permiten monitorear flujos de comunicacion entre host y la red. Esta herramienta clasifica los paquetes obtenidos como posibles pasos en el ciclo de vida de la botnet y realiza perfiles de infeccion que pueden ayudar al usuario a saber si el equipo ha sido comprometido o no

<img src="./Resources Proyecto de grado/bothunter.png">

#### 1.5 Analisis de anomalidad de la red

Se lleva a cabo un analisis general de la red, con el fin de monitorear si exite :

1. Volumen alto en el trafico de red

2. Niveles altos de Latencia

3. Uso de puertos inusuales

## 1. Extraccion de cantidad de paquetes tcp con flag urg activada

Con este algoritmo, se preguntaba si un paquete TCP contenia su bandera URG en 1

```python
def tcp_urg_pck(pkt):
    retorno =0
    try:
            if pkt.transport_layer == 'TCP':
                sumo = False
                for lyr in pkt.layers:
                    flagUrg = str(lyr.get_field_value('tcp.flags.urg'))
                    if flagUrg not in 'None' and flagUrg not in '0':
                        if sumo == False:
                            retorno += 1
                            sumo = True
    except AttributeError:
            pass
    except Exception:
            print ('Error en tcp_urg_packet')
    return retorno
```

```python
#not supported
def remote_pck(capture):
    print ("-------Mining Remote_PCK-------")
    pkts = []
    for pkt in capture:
        try:
            if pkt.ip.dst == 'falta la ip desde donde llegan los pck':
                pkts.append(pkt)
        except AttributeError:
            pass
        except Exception:
            print ('Error en remote_app_packets')
    return pkts
```

## 4. Duracion

Utilizado para estimar la duracion total del pcap. Es la diferencia entre el paquete actual y el primer paquete obtenido durante la captura

```python
#check
def duration(pkt, firstPkt):
    time = float(pkt.frame_info.time_epoch)  -  float(firstPkt.frame_info.time_epoch)
    converted = time * 1000 #Convert seconds to miliseconds
    return converted
```

## 5. AVG Local PCK Rate

```python
#not supported
def avg_local_pck_rate(capture):
    time = duration(capture)
    if time not in 'NA':
        resultado = len(source_pck(capture)) / float(time)
    else:
        resultado = 'NA'
    return resultado
```

## 6. AVG Remote PCK Rate

```python
#not supported
def avg_remote_pck_rate(capture):
    time = duration(capture)
    if time not in 'NA':
        resultado = len(remote_pck(capture)) / float(time)
    else:
        resultado = 'NA'
    return resultado
```

## 13. Dividir el PCAP en PCAPS de 5 minutos

## 23. Cargar y Crear Txt de IPs o Puertos

Con este metodo, la idea es leer los archivos originados con los metodos anteriores, luego de eso pasan a un solo archivo con el fin de minimizar la labor del investigador

```python
def loadAndCreateTxtFull(ruta,comienzo, fin,saveRoute):
    datos = []
    for i in range(comienzo,fin):
        ##ruta = DST_IPs/Benigno_DST_IPs
        r = ruta + str(i)+".0.txt"
        print(r)
        datos.append(r+"\n")
        f = open(r, 'r')
        readlines = f.readlines()
        for y in readlines:
            if(y != ""):
                datos.append(y)

    f.close()
    f = open(saveRoute, 'w')
    for m in range(0,len(datos)):
        if(datos[m] != ""):
         f.write(datos[m])
    f.close()
```

## 24. Cargar y Crear EXCEL de IPs o Puertos

Con este metodo, la idea es leer los archivos originados con los metodos anteriores, luego de eso pasan a un solo archivo excel con el fin de minimizar la labor del investigador

```python
def loadAndCreateExcelFull(ruta,comienzo,fin,saveRoute,type,indicador):
    nameOfCapture = []
    ips = []
    Type = []
    if (indicador == -1):
        for i in range(comienzo, fin):
            ##ruta = DST_IPs/Benigno_DST_IPs
            r = ruta + str(i) + ".0.txt"
            print(r)
            f = open(r, 'r')
            readlines = f.readlines()
            for y in readlines:
                nameOfCapture.append(r)
                ips.append(y)
                Type.append(type)

        f.close()

    else:
            for i in range(comienzo, fin):
            ##ruta = DST_IPs/Benigno_DST_IPs
            r = ruta + str(indicador) + "."+str(i)+".txt"
            #print(r)
            f = open(r, 'r')
            readlines = f.readlines()
            for y in readlines:
                nameOfCapture.append(r)
                ips.append(y)
                Type.append(type)

        f.close()

    df = pd.DataFrame(data={"Name PCAP": nameOfCapture, "IPs": ips, "Type": type})
    writer = ExcelWriter(saveRoute)
    df.to_excel(writer, 'Hoja de datos', index=False)
    writer.save()
```

### B. KNN

Este clasificador, calcula la distancia del elemento nuevo a cada uno de los existentes y ordena dichas distancias de menor a mayor con el fin de ir seleccionando el grupo al que pertenecer. Este grupo sera el de mayor frecuencia con menores distancias

Es un algoritmo de **Aprendizaje supervisado**. El aprendizaje sucede en el mismo momento en el que se prueban los datos de test ( **Lazy learning methods**

**KNN es muy sensible a**:

* La variable **K**, de modo que con valores distintos de *K* podemos obtener resultados muy distintos. Suele fijarse tras un proceso de pruebas con varias instancias

veamos como influye el valor de k :


<img src="./Resources Proyecto de grado/knn.png">


1. Para k =1 , el algoritmo clasificara la esfera con signo + como blanca

2. Para k=2, el algoritmo no tiene criterio para clasificar la espera con signo +

3. Para k>= 3, el algoritmo clasificara la esfera con signo + como negra

* La **metrica** de similitud utilizada, puedo que esta influira fuertemente en las relaciones de cercania que se iran estableciendo en el proeso de construccion del algoritmo.

KNN es utilizado en **DATA Mininig**. Es aplicado en sistemas de recomendacion ( plataformas de contenido digital, procesos de venta cruzada en eCommerce, etc)

### C. Arboles de decision

Los **arboles de decision** , son mapas de posibles resultados de una serie de decisiones relacionadas. Permite a el investigador comparar posibles acciones entre si , segun sus costos, probabilidades y beneficios.

Generalmente comienzan con un unico nodo y luego se ramifica en resultados posibles. Cada uno de los resultados crea nodos adicionales.


Estos arboles pueden poseer diferentes tipos de nodos:

1. **Nodos de probabilidad** : Se representan con un circulo y muestra las probabilidades de ciertos resultados
2.  **Nodos de decision** : representado como un cuadrado, muestra una decision que se tomara
3. **Nodos terminales** : muestra el resultado definitivo de una ruta de decision
