## 4. Let's join Benign and Malign Dataset from Stratosphere

```python
frames = [stratosphere_benign_dataset,stratosphere_malign_dataset]
```

```python
BM_Stratosphere = pd.concat(frames)
```

Let's **shuffle** the dataset:

```python
BM_Stratosphere = BM_Stratosphere.sample(frac=1)
BM_Stratosphere.head()
```
