## 2. Dataset Nomenclatures

1. **Name**: Reference of the time window
2. **Netflows**: Number of netflows in the t.w
3. **First_protocol**: Top 1 of the protocols in the t.w
4. **Second_protocol**: Top 2 of the protocols in the t.w
5. **Third_protocol**: Top 3 of the protocols in the t.w
6. **p1_d**: percentil 25% of the duration
7. **p2_d**: percentil 50% of the duration
8. **p3_d**: percentil 75% of the duration
9. **duration** : sum of all durations in the t.w
10. **max_d**: max number of all duration values in the t.w
11. **min_d**: min number of all duration values in the t.w
12. **#packets**: number of packets in the t.w
13. **Avg_bps**: Average of bits per second
14. **Avg_pps**: Average of packets per second
15. **Avg_bpp**: Average of bytes per package
16. **#Bytes**: Number of bytes in the t.w
17. **#sp**: total of source ports in the t.w
18. **#dp**: total of destination ports in the t.w
19. **first_sp**: top 1 of source ports in the t.w
20. **second_sp**: top 2 of source ports in the t.w
21. **third_sp**: top 3 of source ports in the t.w
22. **first_dp**: top 1 of destination ports in the t.w
23. **second_dp**: top 2 of destination ports in the t.w
24. **third_dp**: top 3 of destination ports in the t.w
25. **p1_ip**: percentil 25% of input packets
26. **p2_ip**: percentil 50% of input packets
27. **p3_ip**: percentil 75% of input packets
28. **p1_ib**: percentil 25% of input bytes
29. **p2_ib**: percentil 50% of input bytes
30. **p3_ib**: percentil 75% of input bytes
31. **type**: Benign/Malign

## 3. Stratosphere's Benign Dataset

### 3.1 Let's Load...

```python
stratosphere_benign_dataset = pd.read_csv(r"C:\Users\Usuario\Documents\Github\PDG\PDG-2\Datasets\Time Window\Originals\Benigns v2\BeningTimeWindowsStratosphere.csv", delimiter = ",")
```

```python
stratosphere_benign_dataset.shape
```

```python
stratosphere_benign_dataset.head()
```

### 3.2 Let's transform the NaN Values...

```python
stratosphere_benign_dataset['First_Protocol'] = stratosphere_benign_dataset['First_Protocol'].replace(np.nan,"None",regex=True)
stratosphere_benign_dataset['Second_Protocol'] = stratosphere_benign_dataset['Second_Protocol'].replace(np.nan,"None",regex=True)
stratosphere_benign_dataset['Third_Protocol'] = stratosphere_benign_dataset['Third_Protocol'].replace(np.nan,"None",regex=True)
```

```python
stratosphere_benign_dataset['second_sp'] = stratosphere_benign_dataset['second_sp'].replace(np.nan,-1,regex=True)
stratosphere_benign_dataset['third_sp'] = stratosphere_benign_dataset['third_sp'].replace(np.nan,-1,regex=True)
```

```python
stratosphere_benign_dataset['second_dp'] = stratosphere_benign_dataset['second_dp'].replace(np.nan,-1,regex=True)
stratosphere_benign_dataset['third_dp'] = stratosphere_benign_dataset['third_dp'].replace(np.nan,-1,regex=True)
```

### 3.3 Let's change dtypes of some columns...

Columns like :

    1. second_sp
    2. third_sp
    3. second_dp
    4. third_dp

Have to be **INTEGER** because they are **PORTS**

```python
stratosphere_benign_dataset['second_sp'] = stratosphere_benign_dataset['second_sp'].astype(np.int64)
stratosphere_benign_dataset['third_sp'] = stratosphere_benign_dataset['third_sp'].astype(np.int64)
```

```python
stratosphere_benign_dataset['second_dp'] = stratosphere_benign_dataset['second_dp'].astype(np.int64)
stratosphere_benign_dataset['third_dp'] = stratosphere_benign_dataset['third_dp'].astype(np.int64)
```

### 3.4 Let's see the dataset...

```python
stratosphere_benign_dataset.info()
```

```python
stratosphere_benign_dataset.describe(include="all").T
```

### 3.5 Let's plot the dataset

```python
fig = plt.figure(figsize=(25,20))
ax = fig.gca()
stratosphere_benign_dataset.hist(ax = ax)
```

```python
invalid = ["First_Protocol","Second_Protocol","Third_Protocol","Type","Name"]
invalid
```

```python
for column in stratosphere_benign_dataset.columns:  
    if column not in invalid:
        fig = plt.figure()
        ax = plt.subplot()
        ax.set_title("Boxplot of %s" %column)
        ax.boxplot(stratosphere_benign_dataset[column])
```

### 3.6 Let's drop some outliers...

#### p1_d

```python
sns.boxplot(y=stratosphere_benign_dataset["p1_d"])
```

```python
values = stratosphere_benign_dataset.iloc[stratosphere_benign_dataset['p1_d'].argsort()[-4:]]
values = values[['p1_d','Name']]
print(values)
```

```python
stratosphere_benign_dataset=stratosphere_benign_dataset[stratosphere_benign_dataset.p1_d<0.017]
```

```python
sns.boxplot(y=stratosphere_benign_dataset["p1_d"])
```

#### first_dp

```python
sns.boxplot(y=stratosphere_benign_dataset["first_dp"])
```

```python
values = stratosphere_benign_dataset.iloc[stratosphere_benign_dataset['first_dp'].argsort()[-4:]]
values = values[['first_dp','Name']]
print(values)
```

```python
stratosphere_benign_dataset=stratosphere_benign_dataset[stratosphere_benign_dataset.first_dp<59370]
```

```python
sns.boxplot(y=stratosphere_benign_dataset["first_dp"])
```

### 3.7 Let's save the cleaned Stratosphere's Benign Dataset

```python
stratosphere_benign_dataset.to_csv(r"C:\Users\Usuario\Documents\Github\PDG\PDG-2\Datasets\Time Window\Cleaned\CleanedStratosphersBenignDataset.csv",sep=',',index=False)
```
