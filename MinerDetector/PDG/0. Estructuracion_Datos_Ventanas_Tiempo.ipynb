# Estructuración de los datos de Ventanas de tiempo

Se adiciona una nueva característica a los conjuntos de datos con registros mineros denominada 'cryptocurrency', la cual me permite saber que tipo de criptomoneda pertenece al proceso de minería. Al conjunto de datos benigno también se le agregó esta columna para poder concatenar todos los registros.

```python
data_benigna['Cryptocurrency'] = 'Does not have'
data_bitcoin['Cryptocurrency'] = 'Bitcoin'
data_bitcash['Cryptocurrency'] = 'Bitcash'
data_litecoin['Cryptocurrency'] = 'Litecoin'
data_eth['Cryptocurrency'] = 'Etherium'
data_monero['Cryptocurrency'] = 'Monero'
data_mineros_no_identificada['Cryptocurrency'] = 'Unidentified'
```

### Concatenación de datos mineros identificados y datos benignos

```python
pure_data_miner_time = pd.concat([data_bitcoin,data_bitcash,data_litecoin,data_eth,data_monero,data_benigna], axis=0)
pure_data_miner_time.head()
```

```python
print(colored(f'Hago un proceso de resetear los indices del dataframe, para que no apunte un mismo indice a varios registros.', attrs=['bold']))
pure_data_miner_time = pure_data_miner_time.reset_index()
```

```python
print(colored(f'Genero un rango de valores', attrs=['bold']))
indices = list(pure_data_miner_time.index)
print(colored(f'El rango de valores se le aplicó aleatoriedad de manera pseudoaleatoria', attrs=['bold']))
random.shuffle(indices)
```

```python
print(colored(f'Modifico los registros del conjunto con los indices aleatorios', attrs=['bold']))
pure_data_miner_time = pure_data_miner_time.reindex(indices)
pure_data_miner_time.head(1)
```

```python
pure_data_miner_time = pure_data_miner_time.drop(['index'], axis = 1)
```

```python
print(colored(f'Guardo el conjunto de datos', attrs=['bold']))
pure_data_miner_time.to_csv('', index=False)
```
