```python
import numpy as np
import pandas as pd
import tensorflow as tf
import swifter
#from model import DeepMALDRL
from tensorflow.keras.models import load_model
```

```python
deepMAL = load_model('deep_mal_v5.h5')
```

```python
print(deepMAL.summary())
```

```python
asproxOld = pd.read_csv('output-TrojanAsproxOld.csv', header=0)
asproxNew = pd.read_csv('output-TrojanAsproxNew.csv', header=0)
electroRAT = pd.read_csv('output-AllElectroRAT.csv', header=0)
```

```python
print("AsproxOld data length: ", len(asproxOld))
print("AsproxNew data length: ", len(asproxNew))
print("ElectroRAT data length: ", len(electroRAT))
```

```python
asproxOld['udps.n_bytes_per_packet'] = asproxOld['udps.n_bytes_per_packet'].swifter.apply(eval)
asproxNew['udps.n_bytes_per_packet'] = asproxNew['udps.n_bytes_per_packet'].swifter.apply(eval)
electroRAT['udps.n_bytes_per_packet'] = electroRAT['udps.n_bytes_per_packet'].swifter.apply(eval)
```

```python
asproxOld= asproxOld.explode('udps.n_bytes_per_packet', ignore_index=True)
asproxNew= asproxNew.explode('udps.n_bytes_per_packet', ignore_index=True)
electroRAT= electroRAT.explode('udps.n_bytes_per_packet', ignore_index=True)
```

```python
#data = list(malware_data['udps.n_bytes_per_packet'])
```

```python
predictionsAsproxOld = deepMAL.predict(list(asproxOld['udps.n_bytes_per_packet']))
predictionsAsproxNew = deepMAL.predict(list(asproxNew['udps.n_bytes_per_packet']))
predictionsElectroRAT = deepMAL.predict(list(electroRAT['udps.n_bytes_per_packet']))
```
